# The infamous missing values NaN

```python
titanic.isnull().head()
```

```python
titanic.isnull().sum()
```

```python
titanic.shape
```

#### There are 891 rows and 11 columns , i.e for cabin out of 891 values 687 values is missing ; now, how to work with clean data with no missing values ; lets drop all rows with missing values 

```python
titanic.dropna(how = 'all').shape  # drop a row if all the values are missing , final dataframe is same since no row is completely NaNs 
```

```python
titanic.dropna(how = 'any').shape  # 891-183 = 708 rows dropped 
```

```python
titanic.dropna(subset=['Name','Survived'] , how = 'any').shape # so all passenger names and survival info is available 
```

```python
titanic.dropna(subset=['Name','Age'] , how = 'any').shape    # some rows dropped in which the passenger age info is not available
```

```python
titanic.shape
```

```python
titanic.Age.value_counts(dropna=False).head()
```
