# Choose all the columns to vlookup from the FROM file to TO file ; both the files should have a matching column which we use to anchor


```python
all_columns = list(df2.columns)
all_columns.remove('Company')
import ipywidgets as widgets
a = widgets.SelectMultiple(options=all_columns,description='columns',disabled=False)
display(a)
```

# display the columns 

```python
list(a.value)
```

# Vlookup to TO file 

```python
%%time
vlookup_columns = list(a.value)
for column in vlookup_columns:
    TO[column]=pd.DataFrame({column:[]})
    TO[column] = TO.Company.map(FROM.set_index("Company")[column].to_dict())
```

# The TO file now has all the columns VLookup from FROM file based on the lookup column "Company"

```python
TO
```

# Export the file

```python
TodaysDate = time.strftime("%Y-%m-%d")
csvfilename = TodaysDate + ".csv"
TO.to_csv(csvfilename, index=False)
    # Proress 
message = widgets.Button(description='process completed',disabled=False,button_style='success', # 'success', 'info', 'warning', 'danger' or ''
tooltip='Click me',icon='check')
display(message)
```
